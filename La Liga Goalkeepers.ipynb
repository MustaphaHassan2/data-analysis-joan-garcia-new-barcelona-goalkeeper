Shot-Stopping Analysis – Who Really Saves the Points?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

df_gk = df[(df['Pos'] == 'GK') & (df['MP'] > 14)].copy()
df_top_20 = df_gk.nlargest(20, 'Saves')

colors = []
for player in df_top_20['Player']:
    if player == 'Joan García':
        colors.append((1.0, 0.0, 0.0)) 
    else:
        colors.append((0.8, 0.7, 0.0)) 

fig, ax = plt.subplots(figsize=(12, 8))
ax.set_facecolor('#0A002A')
fig.patch.set_facecolor('#0A002A')

scatter = ax.scatter(df_top_20['Saves'],
                     df_top_20['Save%'],
                     c=colors, s=100, alpha=0.7, edgecolors='black')

for i, player in enumerate(df_top_20['Player']):
    x = df_top_20['Saves'].iloc[i]
    y = df_top_20['Save%'].iloc[i]
    ax.text(x + .9, y, player, fontsize=7, ha='left', va='center', color='white')

ax.set_title('Shot-Stopping Analysis – Who Really Saves the Points?', color='#AAAAAA', fontsize=16)
ax.set_xlabel('Saves', color='white')
ax.set_ylabel('Save %', color='white')

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.grid(color='white', linestyle='--', alpha=0.1)

tfa_logo = Image.open("D:\\Work\\Data Analysis TFA\\05 Analysing Spain's Emerging Talents in 2025\\logos\\TFA.png")
tfa_logo = tfa_logo.convert("RGBA")
logo_width, logo_height = tfa_logo.size
tfa_logo = tfa_logo.resize((int(logo_width * 0.25), int(logo_height * 0.25)))

fig.figimage(tfa_logo, xo=77, yo=40, alpha=0.8, zorder=3)

plt.tight_layout()
plt.show()
```

Sweeper Keepers – Defensive Actions Outside the Box

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

df_gk = df[(df['Pos'] == 'GK') & (df['MP'] > 14)].copy()
df_top_20 = df_gk.nlargest(20, 'Defensive Actions Outside Penalty Area')

df_top_20 = df_top_20.sort_values(by='Defensive Actions Outside Penalty Area', ascending=True)

colors = ['red' if player == 'Joan García' else '#E6C229' for player in df_top_20['Player']]

fig, ax = plt.subplots(figsize=(12, 8))
ax.set_facecolor('#0A002A')
fig.patch.set_facecolor('#0A002A')

bars = ax.barh(df_top_20['Player'], df_top_20['Defensive Actions Outside Penalty Area'], color=colors, edgecolor='black', alpha=0.8)

for i, player in enumerate(df_top_20['Player']):
    value = df_top_20['Defensive Actions Outside Penalty Area'].iloc[i]
    ax.text(value + 0.5, i, f'{value:.1f}', va='center', color='white', fontsize=7)

ax.set_title('Sweeper Keepers – Defensive Actions Outside the Box', color='#AAAAAA', fontsize=16)
ax.set_xlabel('Defensive Actions Outside Penalty Area', color='white')
ax.set_yticklabels(df_top_20['Player'], color='white', fontsize=7)
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.grid(axis='x', color='white', linestyle='--', alpha=0.1)

tfa_logo = Image.open("D:\\Work\\Data Analysis TFA\\05 Analysing Spain's Emerging Talents in 2025\\logos\\TFA.png").convert("RGBA")
logo_width, logo_height = tfa_logo.size
tfa_logo = tfa_logo.resize((int(logo_width * 0.25), int(logo_height * 0.25)))

fig.figimage(tfa_logo, xo=1070, yo=40, alpha=0.8, zorder=3)

plt.tight_layout()
plt.show()
```
